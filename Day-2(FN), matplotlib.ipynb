```python
#drawing a line in a diagram from position (0,0) to position (6,250):

import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show() # not necessary
```

```python
#drawing a line in a diagram from position (1,3) to position (8,10):

import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,8])
ypoints=np.array([3,10])
plt.plot(xpoints,ypoints, 's')
plt.show() # not necessary
```

```python
#drawing a line in a diagram from position (1,3) to(2,8) then to (6,1) and finally to the position (8,10):

import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,6,8])
ypoints=np.array([-3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show() # not necessary
```

```python
#drawing a line in a diagram from position (1,3) to(2,8) then to (6,1) and finally to the position (8,10):

import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,6,8])
ypoints=np.array([-3,8,1,10])
plt.plot(xpoints,ypoints,'*', color='#CD5C5C')
plt.show() # not necessary
```

```python
#ploting without x-points
ypoints=np.array([2,4,6,8,5,7])
plt.plot(ypoints,'x', color='orangered')
plt.show()
```

```python
# mark each point with a circle
ypoints=np.array([2,4,6,8,5,7])
plt.plot(ypoints,'o:r')
plt.show()
```

```python
# set the size of markers to 20:
ypoints=np.array([2,4,6,8,5,7])
plt.plot(ypoints,':*', ms=20, color='blue')
plt.show()
```

```python
#set the edge colours to red:

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,':o', ms=20, color='grey', mec='r')
plt.show()
```

```python
#set the edge colours to red:

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o', ms=20, mec='r', mfc='b')
plt.show()
```

```python
# use a dotetd line
ypoint=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dotted', marker='o', mfc='r', mec='b')
plt.show()
```

```python
# use a dashed line
ypoint=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dashed', marker='o', mfc='r', mec='b')
plt.show()
```

```python
#drawing of 2 plots

#plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1) #in 1 row, 2 columns, show this plot in 1st column
plt.plot(x,y)

#plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)#in 1 row, 2 columns, show this plot in 2nd column
plt.plot(x,y)

plt.show()
```

```python
# A SIMPLE SCATTER PLOT(for bivariate variables)
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,64,89,67])
plt.scatter(x,y)
plt.show()
```

```python
# drawing 2 plots in the same figure
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,64,89,67])
plt.scatter(x,y)

x=np.array([54,27,48,27,52,17,42,69,74,211,412,94,6])
y=np.array([9,8,87,8,1,6,13,27,4,8,64,9,67])
plt.scatter(x,y)
plt.show()
```

# Bar Plot

```python
# draw 4 bars:

x=np.array(['A','B',"C",'D'])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
# draw 4 bars horizontally:

x=np.array(['A','B',"C",'D'])
y=np.array([3,8,1,10])
plt.barh(x,y, color='orange')
plt.show()
```

```python
# simple histogram, used for having class frequencey, limits
# example: age-population 
x=np.random.normal(170,10,250)
plt.hist(x, color='orange')
plt.show()
```

```python
#a simple pie chart:
y=np.array([10,50,25,15])
plt.pie(y)
plt.show()
```
